  * Iteratively do this until you find the bottom\.

# Neural networks: challenges



* Throwing a bunch of nodes in layers won't necessarily work well:

  * If you have lots of layers\, it will take a long time to train

  * Simply adding more nodes doesn't always increase predictive accuracy

* Use problem knowledge to intelligently construct structure of network

  * What about problems that require very large data?

    * Maybe we want to impose the same weights in different parts e\.g\. "Convolutional Neural Networks"

# Layer types in CNNs



* Convolution layer:

  * Construct one or more "filters" that get passed over the pixel

  * Each filter has a window of say 5x5 pixels\. It moves with some "stride" length over the image\.

  * Need to chose activation function \(sign\, linear\, logistic\)
